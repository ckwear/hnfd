```python
from collections import deque as dq
import pandas as pd
import itertools as it
import math
import copy


maps = list()

r = list()
b = list()
def go(xy, dic):#2가 x축 1이 y축
    
    x = xy[1] + dic
    while True:
        if True: #if 0<=x and x<len(maps[0]):
            #print(xy[0], xy[1], x)
            if maps[xy[0]][x] == 'O'and maps[xy[0]][xy[1]] == 'R':
                return 1
            if maps[xy[0]][x] in '#BR':
                break
        x += dic
    if dic == 1:
        c = -1
    else:
        c = 1
    maps[xy[0]][xy[1]], maps[xy[0]][x+c] = maps[xy[0]][x+c], maps[xy[0]][xy[1]]

    xy[1] = x-dic
    return 0

def upDown(xy, dic):#2가 x축 1이 y축
    x = xy[0] + dic
    while True:
        if maps[x][xy[1]] == 'O'and maps[xy[0]][xy[1]] == 'R':
            return 1
        if maps[x][xy[1]] in '#BR':
            break
        x += dic
    if dic == 1:
        c = -1
    else:
        c = 1
    maps[xy[0]][xy[1]], maps[x+c][xy[1]] = maps[x+c][xy[1]], maps[xy[0]][xy[1]]

    xy[0] = x-dic
    return 0

def dicrect(n):

    if n == 1:
        if r[1] > b[1]:
            rst = go(r, 1)
            rst = go(b, 1)
        else:
            rst = go(b, 1)
            rst = go(r, 1)
    elif n == 2:
        if r[1] <b[1]:
            rst = go(r, -1)
            rst = go(b, -1)
        else:
            rst = go(b, -1)
            rst = go(r, -1)
    elif n == 3:
        if r[0] > b[0]:
            rst = upDown(r,1)
            rst = upDown(b,1)
        else:
            rst = upDown(b,1)
            rst = upDown(r,1)
    else:
        if r[0] < b[0]:
            rst = upDown(r,-1)
            rst = upDown(b,-1)
        else:
            rst = upDown(b,-1)
            rst = upDown(r,-1)        
        

    return rst

        
def main():
    #a = input().split()
    a = '7 7 ####### #...RB# #.##### #.....# #####.# #O....# #######'.split()
    fifo = dq()

    [fifo.append(i) for i in a]
    k = int(fifo.popleft())
    h = int(fifo.popleft())
    for i in range(k):
        l = fifo.popleft()
        if l.find('R') != -1:
            r.append(i)
            r.append(l.find('R'))
            
        if l.find('B') != -1:
            b.append(i)
            b.append(l.find('B'))
            
        maps.append(list(l))
        
        
    
    
    
    
    
def cut():
    min = math.inf
    for c in it.combinations_with_replacement(list(range(1, 5)), 10):
        del r[:]
        del b[:]
        del maps[:]
        [fifo.append(i) for i in a]
        fifo.popleft()
        fifo.popleft()
        for i in range(k):
            l = fifo.popleft()
            if l.find('R') != -1:
                r.append(i)
                r.append(l.find('R'))

            if l.find('B') != -1:
                b.append(i)
                b.append(l.find('B'))
            maps.append(list(l))
        #print(r[0], r[1], b[0], b[1])
        #print(pd.DataFrame(maps, index=list(range(1,k+1)), columns=list(range(1,k+1))), end="\n\n\n\n\n")
        for i,v in enumerate(list(c)):
            
            print(i, v, end="\n")
            print(pd.DataFrame(maps, index=list(range(1,k+1)), columns=list(range(1,k+1))), end="\n\n\n\n\n")
            if dicrect(v) == 1:
                
                if min > i:
                    min = i
                break
            
                
    print(min+1)

        
    
if __name__=="__main__":
    main()
```

```python
from collections import deque
import pandas as pd
import math
maps = []
q = deque()
class Point():
    def __init__(self,ry, rx, by, bx, count):
        self.rx = rx
        self.ry = ry
        self.bx = bx
        self.by = by
        self.count = count
        


def BFS(pt):
    dx = [1, -1, 0, 0]
    dy = [0, 0, 1, -1]
    print(pt.ry, pt.rx, pt.by, pt.bx)
    visited= [[[[0]*10]*10]*10]*10
    
    visited[pt.ry][pt.rx][pt.by][pt.bx] = 1
    q.append(pt)
    while q:
        
        curn = q.popleft()
        if maps[curn.ry][curn.rx] == 'O' or maps[curn.by][curn.bx] == '0':
            return q.popleft().count
        print(1)
        for dic in range(4):
            next_ry = curn.ry
            next_rx = curn.rx
            next_by = curn.by
            next_bx = curn.bx
            
            while True:
                if maps[next_ry][next_rx] != '#' and maps[next_ry][next_rx] != 'O':
                    next_ry += dy[dic]
                    next_rx += dx[dic]
                else:
                    if maps[next_ry][next_rx] == '#':
                        next_ry -= dy[dic]
                        next_rx -= dx[dic]    
                    break
                    
            while True:
                if maps[next_by][next_bx] != '#' and maps[next_by][next_bx] != 'O':
                    next_by += dy[dic]
                    next_bx += dx[dic]
                else:
                    if maps[next_by][next_bx] == '#':
                        next_by -= dy[dic]
                        next_bx -= dx[dic]    
                    break
            if next_ry == next_by and next_rx == next_bx:
                if maps[next_ry][next_rx] != 'O':
                    red_dist = abs(next_ry-curn.ry) + abs(next_rx-curn.rx)
                    blue_dist = abs(next_by-curn.by) + abs(next_bx-curn.bx)
                    if red_dist>blue_dist: next_ry -= dy[dic]
                    else: next_by -= dy[dic]
        
            if visited[next_ry][next_rx][next_by][next_bx] == 0:
                print(pd.DataFrame(maps, index=list(range(1, 8)),columns=list(range(1, 8))))
                q.append(Point(next_ry, next_rx, next_by, next_bx, curn.count+1))
                visited[next_ry][next_rx][next_by][next_bx] = 1
            
                
        
        
def main():
    #a = input().split()
    a = '7 7 ####### #...RB# #.##### #.....# #####.# #O....# #######'.split()
    fifo = deque()
    
    [fifo.append(i) for i in a]
    k = int(fifo.popleft())
    h = int(fifo.popleft())
    for i in range(k):
        l = fifo.popleft()
        if l.find('R') != -1:
            ry = i
            rx = l.find('R')
            
        if l.find('B') != -1:
            by = i
            bx = l.find('B')
            
        maps.append(list(l))
    
    BFS(Point(ry, rx, by, bx, 0))
    
if __name__=="__main__":
    main()
```

```python
class node(object):
    def __init__(self, data = None):
        self.left = None
        self.right = None
        self.data = data

    def __str__(self):
        return str(self.data)
    def setData(self, data):
        self.data = data

class Stack(object):
    def __init__(self):
        self.head = None
        self.count = 0
    def push(self, n):
        if self.head == None:
            self.head = node()
            self.head.left = n
            self.count += 1
        else:
            curn = self.head
            while curn.right is not None:
                curn = curn.right
            curn.right = node()
            curn.right.left = n
            self.count += 1
    def pop(self):
        prevn = None
        curn = self.head
        if self.isEmpty():
            return
        if self.count == 1:
            self.count = 0
            self.head = None
            return curn.left
        while curn.right is not None:
            prevn = curn
            curn = curn.right
        prevn.right = None
        self.count -= 1     
        return curn.left
    
    def peek(self):
        curn = self.head
        if self.isEmpty():
            return
        while curn.right:
            curn = curn.right
        return curn.left
    
    def isEmpty(self):
        if self.head is None:
            return True
        else:
            return False
        
    def print(self):
        curn = self.head
        while curn:
            print(curn.left.data)
            curn = curn.right
```
